```python
# observacion de datos nulos
nulos = base.isnull().sum()
tipos = base.dtypes
show_me = pd.DataFrame ()
show_me['tipos de dato'] = tipos
show_me['cantidad de nulos'] = nulos
show_me['porcentaje_nulos'] = round(nulos*100/base.shape[0],1)
show_me.sort_values(by='porcentaje_nulos',ascending=False)


```

```python
#### dropeo de columnas con mas del 80 % de datos nulos
base_sn = base.loc[:,(base.isnull().sum() <= 399)]
```

```python
def valores_unicos(base):
    tabla = pd.DataFrame()
    for i in range(1,len(base.columns)):
        col = base.columns[i]
        if len(base[col].unique()) <= 2:
             row = pd.DataFrame([[col,base[col].unique()]],columns=["columna","valores"])
             tabla= pd.concat([row,tabla])
    return tabla
```

```python
valores_unicos(base_sn)
```

```python
#se dropean las columnas que contengan un solo valor
base_valores = base_sn.drop(["initiationType","language","tender/documents/0/language"],axis=1)
```
